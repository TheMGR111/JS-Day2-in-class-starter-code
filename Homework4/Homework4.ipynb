```python
# Task 3 response (code):
import pandas as pd
import io
import re

# ---------------------------------------------------------
# 1. LOAD FINANCIAL DICTIONARY (Loughran-McDonald)
# ---------------------------------------------------------
# Using a stable raw GitHub URL since the original is often 404
lm_url = "https://raw.githubusercontent.com/Microkiller/Dic/master/LoughranMcDonald_MasterDictionary_2018.csv"

print("Downloading Financial Dictionary...")
response = requests.get(lm_url)

if response.status_code == 200:
    # Load into DataFrame
    df_lm = pd.read_csv(io.StringIO(response.content.decode('utf-8')))

    # Clean column names (remove extra whitespace)
    df_lm.columns = df_lm.columns.str.strip()

    # Create the 2 exhaustive arrays (using Sets for 100x faster lookup speed)
    # We filter where the dictionary flags the word as non-zero for that sentiment
    positive_financial_words = set(df_lm[df_lm['Positive'] > 0]['Word'].str.lower())
    negative_financial_words = set(df_lm[df_lm['Negative'] > 0]['Word'].str.lower())

    print(f"Dictionary Loaded: {len(positive_financial_words)} positive words, {len(negative_financial_words)} negative words.")

else:
    print("Failed to download dictionary.")
    positive_financial_words = set()
    negative_financial_words = set()

# ---------------------------------------------------------
# 2. PERFORM BAG OF WORDS ANALYSIS
# ---------------------------------------------------------
print("\n--- Bag of Words Analysis Results ---")
print(f"{'Ticker':<10} | {'Total Words':<12} | {'Pos Words':<10} | {'Neg Words':<10} | {'Net Sentiment':<15}")
print("-" * 70)

results_data = []

# Loop through the transcripts loaded in Task 1
for ticker, text in transcripts.items():
    # Pre-processing:
    # 1. Lowercase
    # 2. Regex to remove punctuation (keeps only letters and spaces)
    # 3. Split into tokens
    clean_text = re.sub(r'[^\w\s]', '', text.lower())
    tokens = clean_text.split()

    # Counters
    pos_count = 0
    neg_count = 0

    # Bag of Words Matching
    for word in tokens:
        if word in positive_financial_words:
            pos_count += 1
        elif word in negative_financial_words:
            neg_count += 1

    # Calculate Net Sentiment (Simple: Positive - Negative)
    net_sentiment = pos_count - neg_count
    total_words = len(tokens)

    # Store for optional detailed viewing later
    results_data.append({
        'Ticker': ticker,
        'Positive': pos_count,
        'Negative': neg_count,
        'Net': net_sentiment
    })

    # Print row
    print(f"{ticker:<10} | {total_words:<12} | {pos_count:<10} | {neg_count:<10} | {net_sentiment:<15}")
```

## Task 4: Embeddings

For each transcript, use an embedding-based approach to try to distinguish the positive from negative earnings announcements. Are there key phrases or sentences you find have a strong similarity to the positive earnings announcements and a low similarity to the negative earnings announcements (or vice versa)? If so, what are they? Show everything you tried and the similarity scores you found. Highlight the best sentences/phrases.

```python
# Task 4 response (code):
# Task 4: Embedding-based Sentiment Analysis
from sentence_transformers import SentenceTransformer, util
import torch
import re

# 1. SETUP: Load the Pre-trained Model
# 'all-MiniLM-L6-v2' is a small, fast model optimized for semantic similarity
print("Loading Sentence Transformer Model (this may take a moment)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. DEFINE ANCHORS
# These are the "ideal" sentences we want to compare against.
positive_anchor = "We reported record revenue growth, margin expansion, and strong demand."
negative_anchor = "We experienced a decline in sales, margin pressure, and headwinds affecting our outlook."

# Convert anchors to embeddings
anchor_embeddings = model.encode([positive_anchor, negative_anchor])
pos_anchor_emb = anchor_embeddings[0]
neg_anchor_emb = anchor_embeddings[1]

# 3. ANALYSIS FUNCTION
def analyze_embeddings(transcripts, top_n=3):
    print(f"\n{'Ticker':<8} | {'Best Sentence (Pos Match)':<60} | {'Score':<6}")
    print("-" * 90)

    # Dictionary to store overall sentiment scores for the whole document
    doc_scores = {}

    for ticker, text in transcripts.items():
        # A. Split text into sentences (Simple regex split)
        # We replace newlines with spaces to keep sentences clean
        cleaned_text = text.replace('\n', ' ')
        sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', cleaned_text)

        # Filter out very short sentences (likely headers or noise)
        sentences = [s.strip() for s in sentences if len(s.split()) > 5]

        # B. Encode all sentences in the transcript
        # (This converts text to semantic vectors)
        sentence_embeddings = model.encode(sentences)

        # C. Calculate Similarity Scores
        # Compare every sentence to the Positive Anchor
        pos_sims = util.cos_sim(sentence_embeddings, pos_anchor_emb)
        # Compare every sentence to the Negative Anchor
        neg_sims = util.cos_sim(sentence_embeddings, neg_anchor_emb)

        # D. Find the "Smoking Guns"
        # We want sentences that are VERY close to positive and NOT close to negative
        # Net Score = Similarity to Positive - Similarity to Negative
        net_scores = pos_sims - neg_sims

        # Find the index of the highest scoring sentence (Most Positive)
        best_idx = torch.argmax(net_scores).item()
        best_sentence = sentences[best_idx]
        best_score = net_scores[best_idx].item()

        # Find the index of the lowest scoring sentence (Most Negative)
        worst_idx = torch.argmin(net_scores).item()
        worst_sentence = sentences[worst_idx]
        worst_score = net_scores[worst_idx].item()

        # Print the "Most Positive" finding for this ticker
        # Truncate sentence for display
        display_sent = (best_sentence[:57] + '...') if len(best_sentence) > 57 else best_sentence
        print(f"{ticker:<8} | {display_sent:<60} | {best_score:.4f}")

        # Save detailed data for specific inspection later if needed
        doc_scores[ticker] = {
            "most_positive": (best_sentence, best_score),
            "most_negative": (worst_sentence, worst_score)
        }

    return doc_scores

# 4. RUN THE ANALYSIS
# Ensure 'transcripts' is the dictionary from Task 1
results = analyze_embeddings(transcripts)

# 5. DEEP DIVE: Inspect the Negative Findings
print("\n\n--- Deep Dive: Key Negative Phrases Detected ---")
for ticker, data in results.items():
    sent, score = data['most_negative']
    print(f"\n[{ticker}] Similarity to Negative Anchor: {abs(score):.4f}")
    print(f"\"{sent}\"")
```

```python
# Task 6 response (code):
# ==========================================
# 0. LIBRARIES & SETUP
# ==========================================
!pip install -q -U google-genai sentence-transformers pandas

import requests
import pandas as pd
import json
import time
import re
import torch
from sentence_transformers import SentenceTransformer, util
from google import genai
from google.genai import types
from pydantic import BaseModel, Field

# !!! PASTE YOUR GOOGLE API KEY HERE !!!
GOOGLE_API_KEY = "AIzaSyBWDDafb7E-VyfcHub2-3cWDS1JQDGhFz0"

# 1. LOAD THE TRANSCRIPT (HOG)
url = "https://raw.githubusercontent.com/travlake/ai-investments-course/main/Homework4/HOG_Q4_2025.txt"
print("Downloading HOG Transcript...")
response = requests.get(url)

if response.status_code == 200:
    hog_text = response.text
    print(f"Success! Loaded {len(hog_text)} characters.")
else:
    raise Exception(f"Failed to download transcript. Status: {response.status_code}")

# ==========================================
# PART A: EMBEDDING ANALYSIS (Task 4 Logic)
# ==========================================
print("\n--- Part A: Embedding Analysis (Semantic Search) ---")

# Load Model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define Anchors (The "Ideal" Positive and Negative sentences)
positive_anchor = "We reported record revenue growth, margin expansion, and strong demand."
negative_anchor = "We experienced a decline in sales, margin pressure, and headwinds affecting our outlook."

# Compute Anchor Embeddings
pos_anchor_emb = model.encode(positive_anchor)
neg_anchor_emb = model.encode(negative_anchor)

# Split HOG Transcript into Sentences
cleaned_text = hog_text.replace('\n', ' ')
sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', cleaned_text)
sentences = [s.strip() for s in sentences if len(s.split()) > 5]

# Encode Transcript Sentences
sentence_embeddings = model.encode(sentences)

# Calculate Similarity
pos_sims = util.cos_sim(sentence_embeddings, pos_anchor_emb)
neg_sims = util.cos_sim(sentence_embeddings, neg_anchor_emb)
net_scores = pos_sims - neg_sims

# Find Best Positive Match
best_idx = torch.argmax(net_scores).item()
best_sent = sentences[best_idx]
best_score = net_scores[best_idx].item()

# Find Best Negative Match (The "Smoking Gun")
worst_idx = torch.argmin(net_scores).item()
worst_sent = sentences[worst_idx]
worst_score = net_scores[worst_idx].item()

# Display Embedding Results
print(f"\nMetric | Sentence Found | Score")
print("-" * 80)
print(f"Most Positive | {best_sent[:60]}... | {best_score:.4f}")
print(f"Most Negative | {worst_sent[:60]}... | {worst_score:.4f}")

print("\nFull Negative Sentence found:")
print(f"\"{worst_sent}\"")


# ==========================================
# PART B: LLM ANALYSIS (Task 5 Logic)
# ==========================================
print("\n\n--- Part B: LLM Analysis (Gemini 2.0 Flash) ---")

client = genai.Client(api_key=GOOGLE_API_KEY)

# 1. Update Schema to match your "Best Prompt" metrics
class HogAnalysis(BaseModel):
    guidance_trend: int = Field(..., description="Score 1-5: 1=Lowered/Withdrawn, 5=Raised")
    demand_signal: int = Field(..., description="Score 1-5: 1=Cancellations/Churn, 5=Accelerating")
    inventory_health: int = Field(..., description="Score 1-5: 1=Bloated/Channel Stuffing, 5=Lean")
    executive_tone: int = Field(..., description="Score 1-5: 1=Defensive, 5=Confident")
    bear_case_summary: str = Field(..., description="One sentence summary of the bear case")

# 2. The System Prompt (Your "Best Prompt")
system_instruction = """
Role: You are a Quantitative Financial Analyst. Your job is to convert qualitative text into specific quantitative risk scores.
Task: Analyze the transcript and extract the following structured metrics. Do not provide prose; provide only the data.

Metrics to Extract:
Guidance_Trend: (Score 1-5) 1=Lowered/Withdrawn, 3=Reaffirmed, 5=Raised.
Demand_Signal: (Score 1-5) 1=Cancellations/Churn, 3=Stable, 5=Accelerating/Record Wins.
Inventory_Health: (Score 1-5) 1=Bloated/Channel Stuffing, 5=Lean/Supply constrained.
Executive_Tone: (Score 1-5) 1=Defensive/Evasive, 5=Confident/Direct.
"""

try:
    # 3. Call the API
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=f"{system_instruction}\n\nTRANSCRIPT:\n{hog_text}",
        config=types.GenerateContentConfig(
            response_mime_type="application/json",
            response_schema=HogAnalysis,
            temperature=0.1
        )
    )

    # 4. Process Results
    data = json.loads(response.text)
    data['Ticker'] = "HOG"

    # Create DataFrame
    df = pd.DataFrame([data])

    # Clean up column names for display
    cols = ["Ticker", "bear_case_summary", "guidance_trend", "demand_signal", "inventory_health", "executive_tone"]
    df = df[cols]
    df.columns = ["Ticker", "Bear Case Summary", "Guidance (1-5)", "Demand (1-5)", "Inventory (1-5)", "Tone (1-5)"]

    # Display
    print("\nLLM Quantitative Scores:")
    try:
        from IPython.display import display
        display(df)
    except ImportError:
        print(df.to_string(index=False))

except Exception as e:
    print(f"LLM Error: {e}")
```
